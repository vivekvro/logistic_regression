```python
import pandas as pd
import numpy as np
df = pd.read_csv('term_insurance_data.csv')
```

```python
df['bought insurance'].replace({"no":'0',"yes":'1'},inplace=True)
```

```python
df
```

```python
import matplotlib.pyplot as plt
plt.scatter(x=df['age'],y=df['bought insurance'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(df[['age']],df['bought insurance'],test_size=0.2,random_state=76)
```

```python
print(len(xtrain),len(xtest),len(ytrain),len(ytest))
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(xtrain,ytrain)
```

```python
lr.predict(xtest)
```
